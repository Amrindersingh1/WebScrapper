```python
from urllib.request import urlopen;
from bs4 import BeautifulSoup;
import json;
```

```python
# open the url
url = "http://mylanguages.org/punjabi_alphabet.php";
html = urlopen(url);
```

```python
#Create Beautiful soup object from html
soup = BeautifulSoup(html, 'lxml');
```

```python
# Get the title
title = soup.title
print(title)
```

```python
#find our words table
table = soup.find_all('table')[0]
```

```python
#python dictionary for storing table results
data = []
rows = table.find_all('tr');
```

```python
#parse the rows
for i in range(1,42):
    dict = {}
    columns = rows[i].find_all('td')
    dict['word'] = columns[0].get_text()
    dict['name'] = columns[1].get_text()
    dict['name'] = dict.get('name') + " (" + columns[2].get_text() + ")"
    dict['description'] = columns[3].get_text()
    data.append(dict)
    i += 1
        
```

```python
#print data
print(json.dumps(data))
```
